# Frequency Distribution

How can we automatically idetify the words of a text that are most information about the topic and genre of the text? We can use **FreqDist** class of nltk library, to get the frequency of each vocabulary item in the text.

```python
from nltk import FreqDist, book
```

```python
fdist1 = FreqDist(book.text1)
fdist1
```

FreqDict class return the dictionary containing keys (tokens) and values (frequency of words).

```python
vocabulary = fdist1.keys()
vocabulary
```

```python
fdist1['Moby']
```

We can generate a commulative frequency plot, for *n* number of words.

```python
fdist1.plot(10, cumulative=True)
```
