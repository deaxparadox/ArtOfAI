```python
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
acutal = np.array(['Dog', 'Dog', "Dog", "Not Dog", "Dog", "Not Dog", "Dog", 'Dog', "Not Dog", "Not Dog"])
predicted = np.array(['Dog','Not Dog','Dog','Not Dog','Dog','Dog','Dog','Dog','Not Dog','Not Dog'])
```

```python
# compute the confusion matrix
cm = confusion_matrix(acutal, predicted)
```

```python
sns.heatmap(cm, annot=True, xticklabels=['Dog', "Not Dog"], yticklabels=["Dog", "Not Dog"])
plt.ylabel("Actual", fontsize=13)
plt.title("Confusion matrix", fontsize=17, pad=20)
plt.gca().xaxis.set_label_position("top")
plt.xlabel("Prediction", fontsize=13)
plt.gca().xaxis.tick_top()

plt.gca().figure.subplots_adjust(bottom=0.2)
plt.gca().figure.text(0.5, 0.05, "Prediction", ha="center", fontsize=13)
plt.show()
```

```python
# classifications report based on confusion metrics

print(classification_report(acutal, predicted))
```
