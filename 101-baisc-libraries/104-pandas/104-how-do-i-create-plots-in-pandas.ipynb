```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fifa = pd.read_csv(
    "fifa.csv",
    index_col=0,
    parse_dates=True
)
```

```python
fifa.head()
```

The usage of the `index_col` and `parse_dates` parameters of the `read_csv` function to define the first (0th) column as index of the resulting `DataFrame` and convert the dates in the function to the `Timestamp` objects, respectively.
