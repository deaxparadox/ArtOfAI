# How do I read and write tabular data?

I want to analyze the data in the csv file.

```python
spotify = pd.read_csv(SPOTIFY_DATA)
```

pandas provides the `read_csv()` function to read data stored as a csv file into a pandas `DataFrame`. pandas supports many different file formats or data sources out of the box (csv, excel, sql, json, parquet, ...), each of them with the prefix `read_*`.

Make sure to always have to check on the data after reading in the data. When displaying a `DataFrame`, the first and last 5 rows will be shown be default:

```python
spotify
```

I want to see the first 8 rows of a pandas DataFrame.

```python
spotify.head(8)
```

To see the first N rows of a `DataFrame`, use the `head()` method with the required number of rows (in this case 8) as argument.

To see the N rows from the last, use the `tail()`

```python
spotify.tail(10)
```

A check on how pandas interpreted each of the column data types can be done by requesting the pandas `dtypes` attribute:

```python
spotify.dtypes
```

For each of the columns, the used data type is enlisted. The data types in this `DataFrame` are integers (`int64`), floats (`float64`) and string (`object`).

### If you want to data as a spreadsheet

```python
# it will save the data to spreadsheet (excel)

spotify.to_excel("spotify.xlsx", sheet_name="spotify", index=False)
```

Whereas `read_*` function are used to read data to pandas, the `to_*` methods are used to store data. The `to_excel()` method stores the data as an excel file.

In the example here, the `sheet_name` is named *passengers* instead of the default *Sheet1*. By setting `index=False` the row index labels are not saved in the spreadsheet.

```python
spotify_excel = pd.read_excel("spotify.xlsx", sheet_name="spotify")
```
