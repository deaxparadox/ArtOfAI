# Breast Cancer Classification with a simple Neural Network (NN)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_data = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_data)
print("------------------------DATA----------------------")
print(breast_cancer_data.data)
print("-------------------FEATURES NAME------------------")
print(breast_cancer_data.feature_names)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_data.data, columns=breast_cancer_data.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the `target` column to the data frame
data_frame['label'] = breast_cancer_data.target
```

```python
# print the last 5 rows of the dataframe
data_frame.tail()
```

```python
# number of rows and colums in the dataset
data_frame.shape
```

```python
# getting some information about the data
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# statistical measures about the data
data_frame.describe()
```

```python
# checking the distribution of Target Variable
data_frame['label'].value_counts()
```

1 --> Benign

0 --> Malignant

```python
# total rows
data_frame[data_frame['label'] == 1.0]['label'].count()
```

Separating the features and target

```python
X = data_frame.drop(columns=['label'], axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_std = scaler.fit_transform(X_train)
```

```python
X_test_std = scaler.transform(X_test)
```

Building the Neural Network

```python
# importing tensorflow and keras
import tensorflow as tf
from tensorflow import keras
```

```python
tf.random.set_seed(3)
```

```python
# setting up the layers of Neural Networks

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30, )),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# training the Neural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title("model_accuracy")
plt.ylabel("accuracy")
plt.xlabel("epoch")

plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel("epoch")

plt.legend(["training data", "validation data"], loc='upper right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
