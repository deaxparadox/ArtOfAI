```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DATASET = r"C:\Users\91930\Documents\GITHUB\ArtOfAI\dataset\titanic.csv"
```

```python
titanic = pd.read_csv(DATASET)
```

```python
# 
titanic.head()
```

```python
# getting all NaN values
titanic_notna = titanic.isna()
titanic_notna.head()
```

```python
# update missing value
titanic = titanic.interpolate()
titanic.head()
```

```python
# replacing  male: 1, female: 2
titanic = titanic.replace({"Sex": {"male": 1, "female": 2}}, inplace=False)
titanic.head(2)
```

```python
# droping string columns
titanic = titanic.drop(["Name", "Ticket", "Cabin", "Embarked"], inplace=False, axis=1)
titanic.head(3)
```
